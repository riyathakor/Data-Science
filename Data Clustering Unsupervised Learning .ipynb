```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
```

```python
x,y = load_iris(return_X_y=True)
```

```python
x
```

```python
y
```

```python
sse = [] 
for k in range(1,11):
    km = KMeans(n_clusters=k, init='k-means++', max_iter=300, n_init=10, random_state=2)
    km.fit(x)
    sse.append(km.inertia_)
```

```python
r = sns.lineplot(x=range(1,11), y=sse)
r.set(xlabel = "Number of cluster (k)",
      ylabel = "Sum Squared Error",
      title = "Elbow Method")
```

```python
km = KMeans(n_clusters=3, init='k-means++', max_iter=300, n_init=10, random_state=2)
km.fit(x)
```

```python
km.cluster_centers_
```

```python
pred=km.fit_predict(x)
pred
```

```python
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples=200, n_features=2, centers=4, cluster_std=1.8,random_state=101)
```

```python
data
```

```python
plt.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4)
```

```python
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
f, (ax1, ax2) = plt.subplots(1,2, sharey=True,figsize=(10,6))
ax1.set_title('K means')
ax1.scatter(data[0][:,0],data[0][:,1],c=kmeans.labels_,cmap='magma')
ax2.set_title("Original")
ax2.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```

```python
plt.scatter(data[0][:,0],data[0][:,1],c=kmeans.labels_,cmap='rainbow')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],marker='^',color='green')
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
cancer.keys()
```

```python
print(cancer['DESCR'])
```

```python
dk = pd.DataFrame(cancer['data'],columns=cancer['feature_names'])
```

```python
dk
```

```python
dk.head()
```

```python
from sklearn. preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(dk)
```

```python
scaled_data = scaler.transform(dk)
scaled_data
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
pca.fit(scaled_data)
```

```python
x_pca = pca.transform(scaled_data)
```

```python
scaled_data.shape
```

```python
x_pca.shape
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x_pca[:,0],x_pca[:,1], c =cancer['target'],cmap='plasma')
plt.xlabel('First Principal Components')
```

```python
pca.components_
```

```python
dk_comp = pd.DataFrame(pca.components_,columns=cancer['feature_names'])
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(dk_comp,cmap='plasma') 
```

```python
import numpy as np
x=np.arange(10).reshape(5,2)
y=range(5)

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y)

from sklearn.linear_model import LinearRegression
model= LinearRegression()
model.fit(x_train,y_train)

model.predict([[13,14]])
```

```python
import pickle 

with open('LRmodel.pkl','wb')as f:
    pickle.dump(model,f)
```

```python
with open('LRmodel.pkl','rb')as f:
    saved_model=pickle.load(f)
```

```python
saved_model
```

```python
saved_model.predict([[10,11]])
```

```python
import joblib

f1='LR_model.sav'
joblib.dump(model,f1)

joblib_model=joblib.load(f1)

joblib_model.predict([[14,12]])
```
