```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
x
```

```python
y = x**2
```

```python
plt.plot(x,y,'p-')
plt.xlabel('This is x label')
plt.ylabel('This is y label')
plt.title('CHART')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'g*--')
plt.title('CHART')

plt.subplot(1,2,2)
plt.plot(y,x,'y*--')
plt.title('CHART2')
plt.show()
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y, label = "x = name")
axes.plot(y,x, label = "y = Id")
plt.xlim(-1,11)
plt.ylim(-1.5,1.5)
axes.set_xlabel('set x label')
axes.set_ylabel('set y label')
axes.set_title('Name')
axes.legend(loc=1)
plt.show()
```

```python
import pandas as pd
data = [[101, 'Milk',32],
       [102, 'Butter',45],
       [103, 'NoteBooks',50],
       [104,'Maggie',24],
       [105, 'Body Lotion',350]]
df = pd.DataFrame(data, columns=['ProductId','ProductName','ProductPrice'])
df
```

```python
fig, axes = plt.subplots(2,2)
axes[0,0].plot(df.ProductId,df.ProductName, 'r*--')
axes[0,1].plot(df.ProductId,df.ProductPrice, 'm*--')
axes[1,0].plot(df.ProductId,df.ProductName, 'g*--')
axes[1,1].plot(df.ProductId,df.ProductPrice, 'b*--')
```

```python
np.random.seed(3)
x = 4 + np.random.normal(0, 2, 24)
y = 4 + np.random.normal(0, 2, len(x))

sizes = np.random.uniform(45, 80, len(x))
colors = np.random.uniform(15, 80, len(x))

fig, ax = plt.subplots()
ax.scatter(x,y, s=sizes, c = colors, cmap='brg')
ax.set(xlim=(0,8), ylim=(0,8))
```
